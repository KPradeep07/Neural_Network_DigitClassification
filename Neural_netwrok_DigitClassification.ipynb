```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
# Loading data set from keras 
(X_train,y_train), (X_test, y_test)= keras.datasets.mnist.load_data()
```

```python
# we have seen that shape of X_train is (60000, 28,28), hence we need to flatten it in two dimensions for both x_tran, x_test
X_train_flat = X_train.reshape(len(X_train), 28*28)
X_test_flat = X_test.reshape(len( X_test), 28*28)
```

```python
X_train_flat.shape
```

```python
X_train_flat[0]
```

# Now lets build the model with only 2 layers i.e. Input layer and output layer(without any hidden layer)
# Here we have not scaled the data, so lets try to build the model without scaling and lets check the accuracy of the model

```python
# Here we will create two layers
# Input layer of (28*28= 784)
# Output layer of 10 as we need to classify 10 digits from 0 to 9
# as this is classification problem we wil use Sigmoid activation fuction

model= keras.Sequential(
         [keras.layers.Dense(10, input_shape=(784,), activation = "sigmoid")])
          
model.compile(optimizer= "adam",
             loss= "sparse_categorical_crossentropy",
             metrics = ["accuracy"])
          
model.fit(X_train_flat, y_train, epochs = 20)
```

```python
model.evaluate(X_test_flat, y_test)
```

```python
# here we can see that traing and test accuracy is almost similar
```

```python
y_pred = model.predict(X_test_flat)
```

```python
# It can be seen that the predicted first number is 3
np.argmax(y_pred[0])
```

```python
y_test[0]
```

```python
plt.matshow(X_test[0])
```

# from above results we can seee that it has miss classified the first number

```python
np.argmax(y_pred[2])
```

```python
plt.matshow(X_test[2])
```

```python
# To see the digits at each row 
y_pred_lables = [np.argmax(i) for i in y_pred]
```

```python
# lets create confusion  matrix
con_mat= tf.math.confusion_matrix(labels= y_test, predictions=y_pred_lables)
con_mat
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(con_mat, annot= True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# lets try to scale the values and then will try to bulid the model

```python
# Scaling the data, dividing be 255 because the max number is 255 for white
X_train_scaled = X_train/255
X_test_scaled =X_test/255
```

```python
# we have seen that shape of X_train is (60000, 28,28), hence we need to flatten it in two dimensions for both x_tran, x_test
X_train_flat1 = X_train_scaled.reshape(len(X_train_scaled), 28*28)
X_test_flat1= X_test_scaled.reshape(len( X_test_scaled), 28*28)
```

```python
X_train_flat1[0]
```

# Now lets build the model with only 2 layers i.e. Input layer and output layer(without any hidden layer)

```python
# Here we will create two layers
# Input layer of (28*28= 784)
# Output layer of 10 as we need to classify 10 digits from 0 to 9
# as this is classification problem we wil use Sigmoid activation fuction

model_scaled= keras.Sequential(
         [keras.layers.Dense(10, input_shape=(784,), activation = "sigmoid")])
          
model_scaled.compile(optimizer= "adam",
             loss= "sparse_categorical_crossentropy",
             metrics = ["accuracy"])
          
model_scaled.fit(X_train_flat1, y_train, epochs = 20)
```

```python
model_scaled.evaluate(X_test_flat1, y_test)
```

```python
# Here we have obtained good accuracy with scaled data
```

```python
y_pred1 = model_scaled.predict(X_test_flat1)
```

# Creating the model with hiddden layer now

```python
model1= keras.Sequential(
         [keras.layers.Dense(100, input_shape=(784,), activation = "relu"),
          keras.layers.Dense(32, activation = "relu"),
          keras.layers.Dense(10, activation = "sigmoid")
         ])
          
model1.compile(optimizer= "adam",
             loss= "sparse_categorical_crossentropy",
             metrics = ["accuracy"])
          
model1.fit(X_train_flat1, y_train, epochs = 19)
```

```python
model1.evaluate(X_test_flat1, y_test)
```

```python
y_pred2 = model1.predict(X_test_flat1)
```

```python
# To see the digits at each row 
y_pred2_lables = [np.argmax(i) for i in y_pred2]
```

```python
# lets create confusion  matrix
con_mat2= tf.math.confusion_matrix(labels= y_test, predictions=y_pred2_lables)
con_mat2
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(con_mat2, annot= True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
